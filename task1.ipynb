```python
import matplotlib.pyplot as plt


categories = ['Male', 'Female', 'Other']
counts = [50, 45, 5]


plt.figure(figsize=(8, 6))
plt.bar(categories, counts, color=['blue', 'pink', 'gray'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Gender')
plt.show()
```

```python
import matplotlib.pyplot as plt


ages = [22, 25, 30, 45, 30, 35, 29, 40, 33, 31]


plt.figure(figsize=(8, 6))
plt.hist(ages, bins=5, color='blue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


ages = [22, 25, 30, 45, 30, 35, 29, 40, 33, 31]


plt.figure(figsize=(8, 6))
sns.histplot(ages, bins=5, kde=True, color='blue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution with KDE')
plt.show()
```

```python
import matplotlib.pyplot as plt


ages = [22, 25, 30, 45, 30, 35, 29, 40, 33, 31]


plt.figure(figsize=(8, 6))
plt.hist(ages, bins=5, color='blue', edgecolor='black')
plt.yscale('log')  # Apply logarithmic scale to Y-axis
plt.xlabel('Age')
plt.ylabel('Frequency (log scale)')
plt.title('Age Distribution with Logarithmic Scale')
plt.show()
```
